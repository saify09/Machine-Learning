```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
```

```python
data  = pd.read_csv("Bengaluru_House_Data.csv")
data.head()
```

```python
data = data.drop(['availability','location', 'society'],axis=1)
```

```python
data.head(2)
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.head()
```

```python
data['bhk']=data['size'].str.split(" ").str[0].astype(int)
```

```python
data.drop('size',axis=1,inplace=True)
```

```python
data.head(2)
```

```python
data.total_sqft.unique()
```

```python
data[data.total_sqft.str.contains("-")]
```

```python
def transform_sqft(sqft):
    if "-" in sqft:
        v1,v2 = sqft.split("-")
        v_mean =(float(v1)+float(v2))/2
        return v_mean
    elif sqft.isnumeric():
        return float(sqft)
    else:
        return None             
```

```python
data['total_area']=data.total_sqft.apply(transform_sqft)
```

```python
# 5.31Acres,1000Sq. Meter,1574Sq. Yards ,24Guntha,1500Cents,1Grounds, Perch
```

```python
data.drop('total_sqft',axis=1,inplace=True)
```

```python
data
```

```python
data = pd.get_dummies(data,dtype=int)
```

```python
data.total_area = data.total_area.fillna(data.total_area.median())
```

```python
X = data.drop('price',axis=1)
y = data.price
```

```python
data.corr()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.30,random_state=123) 
```

```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression() 
```

```python
mlr.fit(X_train,y_train)
```

```python
from  sklearn.metrics import root_mean_squared_error, mean_absolute_error, r2_score
y_pre_train = mlr.predict(X_train)
root_mean_squared_error(y_train,y_pre_train)
mean_absolute_error(y_train,y_pre_train)
r2_score(y_train,y_pre_train)
```

```python
y_pred_test = mlr.predict(X_test)
r2_score(y_test,y_pred_test)
```
