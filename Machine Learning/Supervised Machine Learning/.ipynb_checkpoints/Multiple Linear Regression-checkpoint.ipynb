```python
import pandas as pd
```

# Weight-Height

```python
data = pd.read_csv("weight-height.csv")
```

```python
data.head()
```

```python
data.Gender = data.Gender.apply(lambda x:1 if x=="Male" else 0)
```

```python
data.corr()
```

```python
X = data.drop("Weight",axis=1)
y = data.Weight
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.30,random_state=123)
```

```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
```

```python
mlr.fit(X_train,y_train)
```

```python
mlr.coef_
```

```python
mlr.intercept_
```

```python
y_pre_train = mlr.predict(X_train)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
mean_squared_error(y_train,y_pre_train) , r2_score(y_train,y_pre_train)
```

```python
print('Variance score: %.2f' % mlr.score(X_train, y_train))
```

```python
mean_absolute_error(y_train,y_pre_train)
```

```python
mean_squared_error(y_train,y_pre_train)
```

```python
import pickle
```

```python
pickle.dump(mlr, open('Multiple_linear_regression_model_Weight_Height.sav', 'wb'))
```

```python
def Y_house_price_of_unit_area_prediction(No,X1_transaction_date,X2_house_age,X3_distance_to_the_nearest_MRT_station,X4_number_of_convenience_stores,X5_latitude,X6_longitude):
    record= pd.DataFrame({"No":[No],"X1_transaction_date":[X1_transaction_date],"X2_house_age":[X2_house_age],"X3_distance_to_the_nearest_MRT_station":[X3_distance_to_the_nearest_MRT_station],"X4_number_of_convenience_stores":[X4_number_of_convenience_stores],"X5_latitude":[X5_latitude],"X6_longitude":[X6_longitude]})
    #print(record)
    return mlr.predict(record)
```

# Students Performance

```python
data4 = pd.read_csv("Student_Performance.csv")
data4.head()
```

```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
```

```python
mlr.fit(X_train,y_train)
```

```python
mlr.coef_
```

```python
mlr.intercept_
```

```python
y_pre_train = mlr.predict(X_train)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
mean_squared_error(y_train,y_pre_train) , r2_score(y_train,y_pre_train)
```

```python
print('Variance score: %.2f' % mlr.score(X_train, y_train))
```

```python
mean_absolute_error(y_train,y_pre_train)
```

```python
mean_squared_error(y_train,y_pre_train)
```

```python
def Performance_Index_prediction(Hours_Studied,Previous_Scores,Extracurricular_Activities,Sleep_Hours,Sample_Question_Papers_Practiced):
    record= pd.DataFrame({"Hours_Studied":[Hours_Studied],"Previous_Scores":[Previous_Scores],"Extracurricular_Activities":[Extracurricular_Activities],"Sleep_Hours":[Sleep_Hours],"Sample_Question_Papers_Practiced":[Sample_Question_Papers_Practiced]})
    print(record)
    return mlr.predict(record)
```

```python
Hours_Studied=float(input("enter num: "))
Previous_Scores=float(input("enter num: "))
Extracurricular_Activities=float(input("enter num should be zero or one: "))
Sleep_Hours=float(input("enter num: "))
Sample_Question_Papers_Practiced=float(input("enter num: "))
print(f"The expected Performance_Index: {(Performance_Index_prediction(Hours_Studied,Previous_Scores,Extracurricular_Activities,Sleep_Hours,Sample_Question_Papers_Practiced)[0])}")
```
