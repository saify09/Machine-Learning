#### Adding grade using conditions 
    - List comprehsion
    - Function
    - Lambda function

```python
result['grade1'] = [ "A*" if p>=90 
                    else "A" if p>=80 
                    else "B" if p>=70 
                    else "C "if p>=60 
                    else "D" if p>=50 
                    else "F" 
                    for p in result.percentage]
result
```

```python
def assign_grade(p):
    if p>=90:
        return "A*"
    elif p>=80:
        return "A"
    elif p>=70:
        return "B"
    elif p>=60:
        return "C"
    elif p>=50:
        return "D"
    else:
        return "F"
```

```python
result['grade2'] = result.percentage.apply(assign_grade)
result
```

```python
result['grade3']=result.percentage.apply(lambda p:"A*" if p>=90 else "A" if p>=80 else "B" if p>=70  else "C "if p>=60 else "D" if p>=50 else "F")
```
