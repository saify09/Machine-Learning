```python
class Student():
    def __init__(self,roll, name, course, email, campus):
        # instance / object attributed
        self.roll = roll
        self.name = name
        self.course = course
        self.email = email
        self.campus = campus 
    def student_profile(self):
        print( f"""Name: {self.name}
         Roll: {self.roll}
         Course: {self.course}
         Email: {self.email}
         Campus: {self.campus}""")
```

```python
s1 = Student(1234,'asadullah','python','asad@gmail.com', 'zait')
```

```python
s2 = Student(1237,'abdul sami','python','sami@gmail.com', 'zait')
```

```python
# Student.student_profile(s1)
s1.student_profile()
```

```python
s2.student_profile()
```

```python
class Doctor():
    def __init__(self, name, specialization, clinic, timing, fee):
        self.name = name
        self.specialization = specialization
        self.clinic = clinic
        self.timing =timing
        self.fee = fee
    def checkup(self):
        print(f"{self.name} is checking the patients")
    def operate(self):
        print(f"{self.name} is operating patients")
    def prescribe_medicine(self):
        print(f"{self.name} is prescribing medicine to patients")
```

```python
d1 = Doctor('Dr. Rustam', 'Eye', 'TariqRoad','7-10', 1000)
```

```python
d2 = Doctor('Dr. Mastana', 'Brain', 'Mahmoodabad','7-10', 2000)
```

```python
d1.checkup()
```

```python
d2.checkup()
```

```python
class Restaurant():
    def __init__(self,restaurant_name,cuisine_type):
        self.restaurant_name =restaurant_name
        self.cuisine_type = cuisine_type
    def describe_restaurant(self):
        print(self.restaurant_name)
        print(self.cuisine_type)
    def open_restaurant(self):
        print("the restaurant is open")
```

```python
r1 = Restaurant('MashaAllah', 'Russian')
```

```python
r2 = Restaurant('Zouq', 'Pakistani')
```

```python
r1.describe_restaurant()
```

```python
r2.describe_restaurant()
```

```python
class User():
    def __init__(self,first_name,last_name,email,contact):
        self.first_name = first_name
        self.last_name =last_name
        self.email= email
        self.contact = contact
    def describe_user(self):
        print(f"""
               {self.first_name} {self.last_name}
               {self.email}
               {self.contact}
               """)
    def greet_user(self):
        print(f"""
              Mr. {self.first_name} {self.last_name}, you are welcome to my portal""")
```

```python
u1 = User('nasir','hussain', 'nasir@gmail.com', '1234567890')
```

```python
u2 = User('umer','khan', 'khanumer@gmail.com', '32344566468')
```

```python
u1.describe_user()
```

```python
u2.describe_user()
```

```python
u1.greet_user()
```

```python
u2.greet_user()
```

```python
class Car():

    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
    def get_descriptive_name(self):
        long_name = str(self.year) + ' ' + self.make + ' ' + self.model
        return long_name.title()
```

```python
my_new_car = Car('audi', 'a4', 2016)
```

```python
print(my_new_car.get_descriptive_name())
```
