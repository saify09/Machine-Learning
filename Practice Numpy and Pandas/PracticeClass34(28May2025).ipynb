# Discretization and Binning:
    Discrete kare continuous values ko

```python
data = pd.read_excel("employee.xlsx")
```

```python
data.head(2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(x=data.Designation.value_counts().index, height=data.Designation.value_counts().values)
```

```python
plt.hist(data.Salary)
```

```python
data.Salary.fillna(data.Salary.median(), inplace=True)
```

```python
data.Salary
```

```python
bins = [0,40000,80000,120000,160000,200000]
labels = ['low', 'medium', 'average', 'high','very high']
```

```python
data['Salary_Category'] = pd.cut(data.Salary, bins=bins, labels=labels)
```

```python
data
```

```python
data.Salary_Category.value_counts()
```

```python
x = data.Salary_Category.value_counts().index
y = data.Salary_Category.value_counts().values
```

```python
plt.bar(x=x,height=y)
```

```python
d = {'Student': ['Aryan', 'Prajjwal', 'Vishakshi', 'Brijkant', 'Kareena'],
        'Marks': [77, 72, 19, 68, 45]}
df = pd.DataFrame(d)

bins = [0, 50, 75, 100]  # Ranges: 0-50, 51-75, 76-100
lab = ['Low', 'Average', 'High']

# Step 3: Use cut() to categorize the marks
df['Category'] = pd.cut(df['Marks'], bins=bins, labels=lab, include_lowest=True)

print(df)
```
